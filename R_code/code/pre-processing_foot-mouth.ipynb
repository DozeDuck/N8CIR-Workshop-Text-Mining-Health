Within text-mining, or any language analysis context, one data set is usually referred to as 'a corpus' while multiple data sets are referred to as 'corpora'. For text-mining, a corpus can be:
- a set of tweets, 
- the full text of an 18th centrury novel,
- the contents of a page in the dictionary, 
- minutes of local council meetings, 
- random gibberish letters and numbers, or
- just about anything else in text format.

Again, these lines of code won't work unless you create a dataframe called 'df' but you can try them all out on foot_mouth_df to see what they do.

```python
df.loc[2]          # Select Row by Index Label
df.loc['Julia']          # Select Row by Index Label as a string
df.loc[[2,3,6]]    # Select Rows by Index Label List
df.loc[[2,3,'Julia']]    # Select Rows by Index Label List that mixes numbers and strings
df.loc[1:5]     # Select Rows by Label Index Range
df.loc[1:'Julia']     # Select Rows by Label Index Range that uses numbers and strings
df.loc[1:5:2]   # Select Alternate Rows with in Index Labels
```

**To sum up the differences:**

* `.loc` works with numbers and strings whereas `.iloc` only works with numbers
* `.loc` will give errors if you search for a label that isn't there but `.iloc` will just access everything up until the end 
> e.g. if you try `df.iloc[:100]` but only have 50 rows, it will just give you all 50 rows
* When using `.loc` you can select Index ranges and Lists via combinations of numbers and strings!
* With `.loc` you need to specify the labels of rows further down but with `.iloc` you can just use minus indexing - great for if you dont know how many rows your DataFrame has!

Aside from this the syntax is the exact same!

### Accessing columns by index and label

How about accessing columns in a dataframe?

It works a lot like accessing rows, with .iloc (the index position) and .loc (the label). 

BUT! Now you need a comma inside the square brackets. 
* BEFORE the comma (or everything if there is no comma) determines what rows to return, and
* AFTER the comma determines the columns to return.



ALSO! For some reason, you cannot use non-string labels when accessing columns. No idea. 

*So yeah it is **really** looking like .loc is only really beneficial for locating rows/columns via string labels - otherwise it just has too many restrictions!*

```python
# Selecting Columns by Index

foot_mouth_df.iloc[:,1]      # Select Column by Index         
foot_mouth_df.iloc[:,[1]]    # Select Column by READABLE Index 
foot_mouth_df.iloc[:,[1,2]]  # Select Column by Index List - NOTE the extra square brackets NEEDED for this!

foot_mouth_df.iloc[:,:2]     # Select Column via a Range (should give you the first 2 columns)
```

## [EXTRA] Regular conditional filtering vs Boolean masking

I think I have mentioned the idea of 'masking' your dataframe before, so I just thought I'd put a little explanation here incase it's of interest to you. This isn't super important so don't worry if you aren't interested :) 

It's merely another way to filter your dataframe and inspect values that interest you.

Neither method is superior, although boolean masking is said to be faster than regular filtering.

```python
# But, we can do the same thing by creating a boolean mask...

missing = new_foot_mouth['Dates'].isna()

# If we inspect this mask, we can see that it has returned a series of True/False values based on the condition
missing
```

### Tilde operator - for regular conditional filtering + boolean masking

Another filtering tip which is quite useful, is learning how to use the '~' tilde operator.
This operator is used to negate the Boolean values in the dataframe, i.e., True becomes False and False becomes True!

This can be quite useful. For instance, let's say we wanted to filter our dataframe by Occupation, but we want to look at every group apart from Group 6.

```python
# Our instinct might be to do something like this...

groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']

base_foot_mouth[base_foot_mouth['Occupation'].isin(groups)]


# The isin() method is another way of applying multiple conditions for filtering

# However, this is a bit laborious..
```

```python
#Instead, we could achieve the result above by simply using the tilde operator in combination with the str.contains method

base_foot_mouth[~base_foot_mouth['Occupation'].str.contains('Group 6')]
```

```python
# It should be noted that you can also use the != operator in this case too

base_foot_mouth[base_foot_mouth['Occupation'] != 'Group 6']
```
